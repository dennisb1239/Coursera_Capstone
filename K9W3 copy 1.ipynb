# List of postal codes of Canada

### Import relevant libraries 

```python
import pandas as pd
import numpy as np
```

### Read the table from the website 

```python
pcl = pd.read_html('http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)
pcc = pcl[0]
pcc
```

### Remove all rows where the Borough is 'Not assigned' 

```python
index_names = pcc[pcc['Borough']=='Not assigned'].index
pcc.drop(index_names, inplace = True)
pcc
```

### Group remaining rows by Postcode and Borough, and merge the rows with same Postcodes

```python
pcg = pcc.groupby(['Postcode','Borough'])['Neighbourhood'].agg(','.join).reset_index()
pcg
```
